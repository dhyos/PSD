# Understanding data

Dalam kerangka kerja CRISP-DM (Cross Industry Standard Process for Data Mining), tahap Data Understanding merupakan langkah kedua setelah Business Understanding. Tujuannya adalah untuk memahami data yang dimiliki secara menyeluruh sebelum masuk ke tahap pemodelan

## 📊 Macam-macam Tipe Data

Data dalam ilmu data dapat diklasifikasikan dari berbagai sudut pandang. Berikut adalah penjelasan tipe data ditinjau dari **sifat**, **sumber**, **representasi**, dan **struktur**:

---

### 1. Ditinjau dari **Sifat Data**

* **Data Kualitatif (Qualitative Data)**

  * Data berbentuk kategori atau deskripsi, tidak berupa angka.
  * Contoh: jenis kelamin (laki-laki/perempuan), warna mobil, merek motor.
* **Data Kuantitatif (Quantitative Data)**

  * Data berbentuk angka, bisa dihitung dan diukur.
  * Dibagi menjadi:

    * **Diskrit (Discrete)** → berupa hitungan, tidak bisa pecahan. Contoh: jumlah unit terjual.
    * **Kontinu (Continuous)** → berupa ukuran, bisa dalam pecahan. Contoh: harga kendaraan, tinggi badan.

---

### 2. Ditinjau dari **Sumber Data**

* **Data Primer**

  * Data diperoleh langsung dari sumber pertama melalui survei, wawancara, observasi, atau eksperimen.
  * Contoh: kuesioner pelanggan PT. Hasjrat Abadi.
* **Data Sekunder**

  * Data diperoleh dari pihak lain atau sumber yang sudah tersedia.
  * Contoh: laporan keuangan perusahaan, data BPS, dataset publik dari Kaggle.

---

### 3. Ditinjau dari **Representasi Data**

* **Data Numerik**

  * Direpresentasikan dalam bentuk angka.
  * Contoh: harga, kuantitas penjualan, umur.
* **Data Kategorikal**

  * Direpresentasikan dalam bentuk kategori/label.
  * Contoh: tipe kendaraan (MPV, SUV, Hatchback).
* **Data Ordinal**

  * Direpresentasikan sebagai kategori tetapi ada tingkatan.
  * Contoh: kepuasan pelanggan (sangat puas, puas, kurang puas).
* **Data Biner**

  * Hanya memiliki dua nilai.
  * Contoh: Ya/Tidak, Aktif/Nonaktif.
* **Data Teks & Multimedia**

  * Representasi dalam bentuk unstructured data.
  * Contoh: ulasan pelanggan (text), foto kendaraan (image), rekaman iklan (video).

---

### 4. Ditinjau dari **Struktur Data**

* **Data Terstruktur (Structured Data)**

  * Tersimpan dalam format tabel dengan baris dan kolom yang jelas.
  * Contoh: database SQL berisi data transaksi.
* **Data Semi Terstruktur (Semi-Structured Data)**

  * Tidak seketat tabel, tetapi masih punya label/tag.
  * Contoh: file XML, JSON, log server.
* **Data Tidak Terstruktur (Unstructured Data)**

  * Tidak punya format tertentu, biasanya data teks bebas atau media.
  * Contoh: komentar media sosial, gambar produk, video iklan.

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Nama file sesuai dataset Kaggle
file_path = "Dataset PT Hasjrat Abadi.csv"

# Load dataset ke pandas dataframe (dengan encoding fallback)
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "raflyahmadzainur/dataset-pt-hasjrat-abadi-lapulu",
    file_path,
    pandas_kwargs={
        "sep":';',
        "encoding": "latin1"  # atau "ISO-8859-1" jika tetap error
    }
)

```

```python
import pandas as pd
from tabulate import tabulate
if "id" not in df.columns:
  df.insert(0, "id", range(1, len(df) + 1))
head_data = df.head(5)
tail_data = df.tail(5)

# Gabungkan dengan pemisah "......."
separator = pd.DataFrame([["...." for _ in df.columns]], columns=df.columns)

# Gabungan
combined = pd.concat([head_data, separator, tail_data], ignore_index=True)

# Tampilkan dalam bentuk tabel
print(tabulate(combined, headers="keys", tablefmt="grid",showindex=False))
```

Menapilkan data dalam bentuk diagram
Proses merupakan proses untuk menampilkan data menjadi diagram sekaligus menetukan median mean dan modus dari setiap label atau fiturnya dan membedakan warna diagram menyesuaikan dengan banyaknya class yang ada.

```python
df.columns = df.columns.str.strip()  # hapus spasi depan & belakang
print(df.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Bersihkan kolom Harga
df["Harga"] = df["Harga"].replace({",": ""}, regex=True).astype(float)

# Parse tanggal
df["Tanggal Pemesanan"] = pd.to_datetime(
    df["Tanggal Pemesanan"],
    dayfirst=True,
    errors="coerce"
)

# ------------------- VISUALISASI -------------------

# 1. Jumlah Kuantitas per Model (Top 15)
plt.figure(figsize=(12,8))
data1 = df.groupby("Model")["Kuantitas"].sum().sort_values(ascending=False).head(15)
sns.barplot(y=data1.index, x=data1.values, palette="viridis", legend=False)
plt.title("Top 15 Jumlah Kuantitas per Model", fontsize=14, weight="bold")
plt.xlabel("Total Kuantitas")
plt.ylabel("Model")
plt.tight_layout()
plt.show()

# 2. Harga Rata-rata per Model (Top 15)
plt.figure(figsize=(12,8))
data2 = df.groupby("Model")["Harga"].mean().sort_values(ascending=False).head(15)
sns.barplot(y=data2.index, x=data2.values, palette="mako", legend=False)
plt.title("Top 15 Harga Rata-rata per Model", fontsize=14, weight="bold")
plt.xlabel("Harga (Rata-rata)")
plt.ylabel("Model")
plt.tight_layout()
plt.show()

# 3. Pesanan per Daerah (Top 15)
plt.figure(figsize=(12,8))
data3 = df.groupby("Tempat Tinggal")["id"].count().sort_values(ascending=False).head(15)
sns.barplot(y=data3.index, x=data3.values, palette="crest", legend=False)
plt.title("Top 15 Jumlah Pesanan per Daerah", fontsize=14, weight="bold")
plt.xlabel("Jumlah Pesanan")
plt.ylabel("Daerah")
plt.tight_layout()
plt.show()

# 4. Trend Penjualan per Hari
plt.figure(figsize=(14,6))
data4 = df.groupby("Tanggal Pemesanan")["Kuantitas"].sum().reset_index()
sns.lineplot(data=data4, x="Tanggal Pemesanan", y="Kuantitas", marker="o", color="steelblue")
plt.title("Trend Penjualan Harian", fontsize=14, weight="bold")
plt.xlabel("Tanggal Pemesanan")
plt.ylabel("Total Kuantitas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Indentifikasi Masalah Data

merupakan proses memahami kualitas data dan memastikan bahwa data yang akan kita analisis atau gunakan dalam model machine learning sudah bersih, lengkap, dan sesuai format yang diharapkan. Hal yang perlu diperhatikan adalah:

Missing Values: Menemukan data yang hilang atau tidak lengkap.

Outliers: Mendeteksi nilai-nilai ekstrem yang mungkin merupakan kesalahan atau anomali.

Duplikasi Data: Mengidentifikasi baris atau entri yang duplikat.

Inkonsistensi: Memeriksa apakah ada ketidaksesuaian dalam format atau nilai data.

```python
import pandas as pd
import matplotlib.pyplot as plt


# --- 1. Identifikasi Masalah Data ---
print("=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Duplikasi ===")
print(f"Jumlah duplikat: {df.duplicated().sum()}")

print("\n=== Statistik Deskriptif ===")
print(df.describe())

# --- 2. Scatter Plot ---

plt.figure(figsize=(12,5))

# Scatter plot: Kuantitas vs Harga
plt.subplot(1,2,1)
plt.scatter(df["Kuantitas"], df["Harga"], alpha=0.5, c="blue")
plt.title("Scatter Plot Kuantitas vs Harga")
plt.xlabel("Kuantitas")
plt.ylabel("Harga")

# Scatter plot: ID Pelanggan vs Harga
plt.subplot(1,2,2)
plt.scatter(df["ID Pelanggan"], df["Harga"], alpha=0.5, c="green")
plt.title("Scatter Plot ID Pelanggan vs Harga")
plt.xlabel("ID Pelanggan")
plt.ylabel("Harga")

plt.tight_layout()
plt.show()
```

Interquartile Range (IQR) adalah ukuran statistik yang digunakan untuk melihat sebaran data di sekitar nilai tengah. IQR dihitung dari selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1).

Rumus:

$$
IQR = Q3 - Q1
$$

Q1 adalah nilai kuartil pertama (25% data terbawah), sedangkan Q3 adalah nilai kuartil ketiga (75% data terbawah).
IQR sering dipakai untuk mendeteksi outlier, dengan menentukan batas bawah dan batas atas.

$$
\text{Batas Bawah} = Q1 - 1.5 \times IQR
$$

$$
\text{Batas Atas} = Q3 + 1.5 \times IQR
$$

Setiap data yang berada di bawah batas bawah atau di atas batas atas dianggap sebagai outlier.
