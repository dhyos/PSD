# Bussiness Understanding

PT. Hasjrat Abadi Lapulu merupakan salah satu outlet dealer resmi kendaraan bermotor Yamaha yang beroperasi di Kelurahan Lapulu, Kendari, Sulawesi Tenggara. Outlet ini diperkenalkan kepada publik melalui acara soft opening pada awal September 2022, dan sengaja hadir untuk mempermudah akses konsumen lokal dalam hal pembelian maupun layanan servis sepeda motor Yamaha, tanpa perlu melakukan perjalanan jauh ke pusat kota Kendari
Dinamika Sultra. Sebagai bagian dari jaringan PT. Hasjrat Abadi, perusahaan ini turut memperluas jangkauan bisnis otomotif di Indonesia Timur, mendukung misi memperkuat layanan, penjualan, dan purna jual Yamaha bagi masyarakat di wilayah tersebut

## Tujuan Bisnis

Tujuan Bisnis yang akan saya analsis nantinya yaitu mengetahui produk apa yang laku terjual, daerah mana saja yang memiliki penjualan paling banyak agar nantinya bisa menjadi pertimbangan perusaan untuk mengembangkan bisnisnya

## Situaasi dan kondisi  

Tentunya perusahaan ini sudah berdirin lama sejak 1 November 1927 di
Manado. hal itu membuat kepercayaan masyarakat pada peruhaan tersebut sangatlah tinggi ditambah lagi dengan fokus perusaan yang bergerak di penjualan otomotif seperti yang kita ketahui bahwasanya bidang otomotif saat ini sangatlah berkembang pesat dan daya minat masyarakat cenderung meningkat. Hal ini yang memungkinkan perusahaan ini dapat berkembang lebih besar lagi di era saat ini harapannya dari hasil analisis nantinya bisa membantu perusahaan untuk mengembangkan bisnisnya
